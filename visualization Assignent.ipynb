```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris #(iris dataset)
from sklearn import decomposition


d = load_iris()  # load data
data_array = d.data  # data to ndarray
print("Array shape before decomposition : ",data_array.shape)  # shape before decomposition
decomposed_array = decomposition.PCA(n_components=3).fit_transform(data_array)  # using PCA to reduce dimension from 4 to 3
print("Array shape after decomposition : ",decomposed_array.shape) # shape after decomposition
# below code is for 3D plot
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(decomposed_array[:,0],decomposed_array[:,1],decomposed_array[:,2],c=d.target) #d.target will give target values.
plt.show()
plt.close()
```
